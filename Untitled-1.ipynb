```python
data = pd.read_csv('4_mkr_credits.csv')
```

```python
# create time since start 
data['given_date'] = pd.to_datetime(data['given_date'])
curent_date = pd.to_datetime('today')
data['time_since_start'] = (curent_date - data['given_date']).dt.days
```

```python
import pandas as pd

# Ensure both columns are datetime objects
data['file_date'] = pd.to_datetime(data['file_date'], errors='coerce')
data['last_end_date'] = pd.to_datetime(data['last_end_date'], errors='coerce')

# Apply lambda function to calculate loan duration
data['loan_duration'] = data.apply(
    lambda row: (row['last_end_date'] - row['file_date']).days if row['is_active'] == 1 else 0, 
    axis=1
)

```

```python
data.to_csv('mkr_credits.csv')
```

```python
data = pd.read_csv('mkr_credits.csv')
data.head()
```

```python
data['given_date'] = pd.to_datetime(data['given_date'])
data['file_date'] = pd.to_datetime(data['file_date'])
data['last_end_date'] = pd.to_datetime(data['last_end_date'])
```

```python

# Calculate the difference in years and store it in the 'year_diff' column
data['year_diff'] = (data['file_date'].dt.year - data['given_date'].dt.year) +1
```

```python
data['encode_label']= data['loan_duration'].apply(lambda x: 1 if x > 0 else ( -1 if x < 0 else 0))
```

```python
data['scale_loan_duration'] = data['loan_duration']  // data['year_diff']
```

```python
df= data[['customer_id', 'given_date','file_date','year_diff','loan_duration','scale_loan_duration', 'encode_label', 'remaining_amount']]
```

```python
df.info()
```

```python
import numpy as np

df_group = df.groupby('customer_id').apply(
    lambda x: (x['loan_duration'] * x['remaining_amount']).sum() / x['remaining_amount'].sum()
    if x['remaining_amount'].sum() != 0 else np.nan
).reset_index(name='WeightedDaysLeftToPay')
```

```python
df_group
```

```python

# Ensure data is numeric
df['loan_duration'] = pd.to_numeric(df['loan_duration'], errors='coerce')
df['remaining_amount'] = pd.to_numeric(df['remaining_amount'], errors='coerce')
df['scale_loan_duration'] = pd.to_numeric(df['scale_loan_duration'], errors='coerce')
df['encode_label'] = pd.to_numeric(df['encode_label'], errors='coerce')
df_group = df.groupby('customer_id').agg(
    total_loan_duration=('scale_loan_duration', 'sum'),
    total_encode_label=('encode_label', 'sum'),
    dayslefttopay=(
        'customer_id',
        lambda x: (
            (df.loc[x.index, 'loan_duration'] * df.loc[x.index, 'remaining_amount'].fillna(0)).sum() /
            df.loc[x.index, 'remaining_amount'].fillna(0).sum()
            if df.loc[x.index, 'remaining_amount'].fillna(0).sum() != 0 else 0
        )
    )
).reset_index()
```
